```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('Invistico_Airline.csv')
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
data.drop(columns=['Customer Type','Type of Travel'],axis=1,inplace=True)
data.head()
```

```python
data['satisfaction']=[1 if value=='satisfied' else 0 for value in data['satisfaction']]
#satisfied==1 and notsatisfied==0
```

```python
data.head()
```

```python
## Male==1, Female==0
## satisfied==1, Eco==0
data['Gender']=[1 if value=='Male' else 0 for value in data['Gender']]
data['Class']=[1 if value=='C' else 0 for value in data['Class']]
```

```python
data.head()
```

```python
plt.hist(data)
```

```python
data.corr()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Arrival Delay in Minutes']
```

```python
data.drop(columns=['Arrival Delay in Minutes'],axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns="satisfaction",axis=1)
y=data["satisfaction"]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
print(score)
print(cm)
print(cr)
```

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
